```python
import statistics

from statistics import mean

list1=[]

list1.append(data2['2006'].mean())

list1.append(data2['2007'].mean())

list1.append(data2['2008' ].mean())

list1.append(data2 [ '2009' ].mean ())

list1.append(data2 [ '2010' ].mean())

print(list1)

m1= mean (list1) 

s1= round (statistics.pstdev(list1), 2)

print("mean for decade 2005-2010 [x1]:",m1)

print("standard deviation for the time period of 2006-2010 [s1]:" ,s1)
```

```python
import statistics

from statistics import mean

list2=[]

list2.append(data2[ '2011' ].mean())

list2.append(data2[ '2012' ].mean())

list2.append(data2[ '2013' ].mean())

list2.append(data2[ '2014' ].mean())

list2.append(data2[ '2015' ].mean())

print(list2)

m2= mean (list2) 

s2= round (statistics.pstdev(list2), 2)

print("mean for decade 2011-2015 [x2]:",m2)

print("standard deviation for the time period of 2011-2015 [s2]:" ,s2)
```

```python
import math
standard_deviation=round(math.sqrt((pow(s1,2)+pow(s2,2))/2),2)
standard_deviation
```

```python
t=round((m2-m1)/(standard_deviation*math.sqrt(0.2+0.2)),2)
t
```

```python
t=2.26
print(t)
if(t>t):
    print("rejection of null hypothesis")
else:
    print("accepting the null hypothesis")
```
