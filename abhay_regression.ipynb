```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.scatter(d1.columns[1:],d1.iloc[10,1:])
plt.xlabel('years')
plt.ylabel('mean')
plt.show()
```

```python
table=pd.DataFrame(d1.iloc[10,1:])
table.columns=['Y']
table['X']=range(1,11)
table['XY']=table['Y']*table['X']
table['X^2']=table['X']**2
sum=pd.DataFrame(table.sum())
sum.columns=['Sigma']
sum
mean=pd.DataFrame(table.mean())
mean.columns=['Mean']
mean
table=pd.concat([table,sum.T,mean.T])
table
```

```python
numerator=table['XY'].loc['Sigma']-table['X'].loc['Sigma']*table['Y'].loc['Sigma']/10
denominator=table['X^2'].loc['Sigma']-table['X'].loc['Sigma']**2/10
b=numerator/denominator
print("b=",b)
```

```python
y_bar=table['Y'].loc['Sigma']/10
x_bar=table['X'].loc['Sigma']/10
a=y_bar-b*x_bar
print("a=",a)
```

```python
def eq_regression(x):
    y=table['Y'].loc['Mean']+b*(x-table['X'].loc['Mean'])
    return(y)
plt.figure(figsize=(10,6))
x_values=[]
y_values=[]
for x in range(11,21):
    x_values.append(x)
    y_values.append(eq_regression(x))
plt.scatter([range(11,21)],y_values)
predictions=pd.DataFrame(y_values,[i+2010 for i in x_values])
predictions.columns=['Predicted access to electricity']
print(predictions)

temp=[]
for i in d1.columns[1:]:
    temp.append(int(i)-2010)
plt.scatter(temp,d1.iloc[10,1:])


x_line=[]
y_line=[]
for x in range(1,21):
    x_line.append(x)
    y_line.append(eq_regression(x))
plt.plot(x_line,y_line)
plt.xlabel('Years')
plt.ylabel('access to electricity')
plt.show()
```
