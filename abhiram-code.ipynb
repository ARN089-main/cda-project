```python
import numpy as np
import sys
n = int(input('Enter order of matrix: ')) 
a=np.zeros((n,n))
print("Enter Matrix Coefficients:")
for i in range(n): 
    for j in range(n):
        a[i][j] = float(input('a['+str(i)+']['+ str(j)+']='))
x= np.zeros((n)) 
print('Enter initial guess vector: ')
for i in range(n):

    x[i] =float (input( 'x['+str(i)+']=')) 
tolerable_error= float(input('Enter tolerable error:'))
max_iteration= int(input("Enter maximum number of steps: "))
lambda_old =1.0
condition= True
step = 1
while condition:

    x= np.matmul(a,x)
    lambda_new=max(abs(x))
    x = x/lambda_new 
    print("\nSTEP %d" %(step))
    print("-----------")
    print("Eigen Value =%0.4f" %(lambda_new))

    print("Eigen Vector: ")

    for i in range(n):

        print("%0.3f\t"%(x[i]))

    step=step + 1 
    if step > max_iteration:

        print('Not convergent in given maximum iteration!')

        break

    error= abs(lambda_new-lambda_old) 
    print("errror="+ str(error))
    lambda_old =lambda_new 
    condition =error>tolerable_error
```

```python
DATA33=DATA3.iloc[6]
DATA33
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.scatter(DATA3.columns[1:],DATA3.iloc[6,1:])
plt.xlabel('years')
plt.ylabel('renewable hydro energy produced in india  in Twh')
plt.show()
```

```python
table=pd.DataFrame(DATA3.iloc[6,1:])
table.columns=['Y']
table['X']=range(1,21)
table['XY']=table['Y']*table['X']
table['X^2']=table['X']**2
sum=pd.DataFrame(table.sum())
sum.columns=['Sigma']
sum
mean=pd.DataFrame(table.mean())
mean.columns=['Mean']
mean
table=pd.concat([table,sum.T,mean.T])
table
```

```python
numerator=table['XY'].loc['Sigma']-table['X'].loc['Sigma']*table['Y'].loc['Sigma']/20
denominator=table['X^2'].loc['Sigma']-table['X'].loc['Sigma']**2/20
b=numerator/denominator
print("b=",b)
```

```python
def eq_regression(x):
    y=table['Y'].loc['Mean']+b*(x-table['X'].loc['Mean'])
    return(y)
plt.figure(figsize=(10,6))
x_values=[]
y_values=[]
for x in range(21,32):
    x_values.append(x)
    y_values.append(eq_regression(x))
plt.scatter([range(21,32)],y_values)
predictions=pd.DataFrame(y_values,[i+1999 for i in x_values])
predictions.columns=['Predicted capacity of hydro power in India in TWH']
print(predictions)

temp=[]
for i in DATA3.columns[1:]:
    temp.append(int(i)-1999)
plt.scatter(temp,DATA3.iloc[6,1:])


x_line=[]
y_line=[]
for x in range(1,37):
    x_line.append(x)
    y_line.append(eq_regression(x))
plt.plot(x_line,y_line)
plt.xlabel('Years')
plt.ylabel('production capacity of hydro power in India in TWH')
plt.show()

```
